```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
RNST = 1373
DF = pd.read_csv('../../datasets/Automobile.csv',low_memory=False,
                 usecols=[
                     'bore','fuel_type','number_of_doors','body_style','engine_location','engine_type','engine_size',
                     'fuel_system','bore','horsepower','peak_rpm','city_mpg','highway_mpg',
                     'compression_ratio','stroke','price'
                 ])
```

```python
X = DF[DF.columns.difference(['price'])]
Y = np.log(DF['price'])
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state=RNST)
```

```python
X_train.head()
```

```python
from catboost import Pool, CatBoostRegressor
```

```python
train_pool = Pool(X_train, 
                  Y_train, 
                  cat_features=[0,4,6,7,8,11])

test_pool = Pool(X_test, 
                  Y_test, 
                  cat_features=[0,4,6,7,8,11])

# del DF,X,Y,X_test,X_train,Y_test,Y_train
```

```python
# model = CatBoostRegressor(iterations=2000,learning_rate=.05,depth=15,eval_metric='MAE',random_state=3020)
# model.fit(X_train,Y_train,cat_features=[],eval_set=(X_test,Y_test),plot=True)
```

```python
model = CatBoostRegressor(iterations=8000,learning_rate=.05,depth=6,eval_metric='MAE',random_state=RNST)
model.fit(train_pool,eval_set=(test_pool))
```

```python
Y_predict = model.predict(test_pool)
```

```python
model.score(test_pool)
```

```python
MAE = mean_absolute_error(Y_test,Y_predict)
np.exp(MAE)
```

```python
r2_score(Y_test,np.exp(Y_predict))
```

```python
with open('catboost_regression_model.m74','wb') as file:
    pickle.dump(model,file)
```
