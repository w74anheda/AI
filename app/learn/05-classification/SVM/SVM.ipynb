```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as PLT
import numpy as np
import pandas as pd
from matplotlib import pyplot as PLT, cm as CMAP
from sklearn import datasets
```

```python
X,Y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=.60)
PLT.scatter(X[:,0],X[:,1],c=Y,s=50,cmap=PLT.cm.coolwarm)
```

```python
X,Y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=.60)
xfit= np.linspace(-1,3.5)
PLT.scatter(X[:,0],X[:,1],c=Y,s=50,cmap=PLT.cm.coolwarm)

for m,b in [(1,0.65),(0.5,1.6),(-0.2,2.9)]:
    PLT.plot(xfit, m*xfit+b, '-k')
    
PLT.xlim(-1,3.5)
# select the line with larg margin with all groups
```

```python
X,Y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=.60)
xfit= np.linspace(-1,3.5)
PLT.scatter(X[:,0],X[:,1],c=Y,s=50,cmap=PLT.cm.coolwarm)

for m,b,d in [(1,0.65,0.33),(0.5,1.6,0.55),(-0.2,2.9,0.2)]:
    yfit = m*xfit+b
    PLT.plot(xfit, yfit, '-k')
    PLT.fill_between(xfit,yfit-d,yfit+d,edgecolor='none',color='#AAAAAA',alpha=0.4)
    
PLT.xlim(-1,3.5)
```

# <font color="orange">Mnist Classification with SVM</font>

```python
DT = datasets.load_digits()
```

```python
img_labels = list(zip(DT.images,DT.target))
for index,(image,label) in enumerate(img_labels[100:104]):
    PLT.subplot(1,4,index+1)
    PLT.imshow(image,cmap=CMAP.gray_r)
    PLT.title(label)
```

```python
DT.images.shape
```

```python
X = DT.images.reshape(len(DT.images),-1)
X.shape
```

```python
Y = DT.target
```

# <font color="orange">SVM algoritm (Support Vector Machine) - hyper plane</font>

```python
from sklearn.svm import SVC
```

```python
svm_classifier = SVC(C=1)
```

```python
svm_classifier.fit(X[:1000],Y[:1000])
```

```python
number = 190
PLT.imshow(DT.images[number],cmap=CMAP.gray_r)
PLT.title(DT.target[number])
svm_classifier.predict([X[number]])[0]
```

```python
NewVector = X[190].copy()
NewVector[35] = 16
NewVector[34] = 16
NewVector[33] = 16
NewVector[25] = 16
NewVector[18] = 16
NewVector[17] = 16

PLT.imshow(NewVector.reshape(8,8),cmap=CMAP.gray_r)
svm_classifier.predict([NewVector])[0]
```

```python
predicted = svm_classifier.predict(X[1000:])
expected = Y[1000:]
```

# <font color="orange">accuracy / precision / recall / f1-score</font>

```python
from sklearn import metrics
```

```python
report = metrics.classification_report(expected,predicted)
print(report)
```

```python
metrics.confusion_matrix(expected,predicted)
```
