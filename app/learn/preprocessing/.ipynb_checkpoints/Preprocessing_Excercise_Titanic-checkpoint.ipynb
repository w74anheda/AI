```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as PLT
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import Binarizer
import seaborn as SB
import matplotlib.cm as CMAP
```

```python
DF = pd.read_csv('../datasets/titanic.csv')
DF.head()
```

# <font color='blue'>Titanic</font>

```python
def label_encoder(DF,column_name):
    label_encoder = LabelEncoder()
    encoded = label_encoder.fit_transform(DF[column_name])
    labels =pd.Series(label_encoder.classes_)
    return labels,encoded
```

```python
def OHC_transformer_single_column(DF,column_name,merged = False):
    label_encoder = LabelEncoder()
    encoded = label_encoder.fit_transform(DF[column_name])
    labels =pd.Series(label_encoder.classes_)
    
    clmt = ColumnTransformer([(column_name,OneHotEncoder(),[0])], remainder='passthrough')
    clmt_fited_data = clmt.fit_transform(DF[[column_name]]) 
    clmt_fited_data = clmt_fited_data if type(clmt_fited_data)== np.ndarray else clmt_fited_data.toarray()  
    NDF = pd.DataFrame(clmt_fited_data,columns=[f'{column_name}_{x}' for x in labels])
    
    if merged is True :
        merged = DF.merge(NDF, left_index=True,right_index=True)
        return merged.drop(columns=[column_name])
    else: 
        return NDF
```

```python
PpDF = DF
```

```python
smi_mode = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
knimp = KNNImputer(n_neighbors=5)
```

```python
DF[['Embarked']] = smi_mode.fit_transform(DF[['Embarked']])
DF[['Cabin']] = smi_mode.fit_transform(DF[['Cabin']])
```

```python
for x in ['Sex','Ticket','Cabin','Embarked']:
    PpDF = OHC_transformer_single_column(PpDF,x,True) 
```

```python
KNNDF = PpDF[PpDF.columns.difference(['Name','PassengerId'])]
PpDF[PpDF.columns.difference(['Name','PassengerId'])] = knimp.fit_transform(KNNDF)
```
